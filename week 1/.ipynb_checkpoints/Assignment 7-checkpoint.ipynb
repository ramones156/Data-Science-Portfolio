## Test 2 - Distribution between price and discount

```python
steam[['name','original_price']].sort_values(by='original_price')
```

As you can see, some of these values are also in other currencies, but for this test it will be fine.

```python
steam["original_price"] = steam["original_price"].str.replace('$','') # remove dollar sign
steam["original_price"] = pd.to_numeric(steam["original_price"], downcast="float", errors='coerce') # convert to float
```

```python
steam[['name','original_price']].sort_values(by='original_price')
```

```python
priceDist = pd.Series(steam.original_price).dropna()
```

```python
priceDist.value_counts().sort_index().plot(kind='bar')
```

The amount of games make it difficult to see, but you can definitely see a shape in this graph
