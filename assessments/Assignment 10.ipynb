```python
vg = pd.read_csv('vgsales.csv')
vg.head()
```

Let's check the bivariate correlation between countries.

```python
corr = vg[['EU_Sales','NA_Sales','JP_Sales','Other_Sales','Global_Sales']].corr()
corr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
sns.scatterplot(x=corr.NA_Sales,y=corr.Global_Sales) # Strongest positive
```

```python
sns.scatterplot(x=corr.Other_Sales,y=corr.JP_Sales) # Weakest
```
